```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

model.classes =[0,]
```

```python
import math
import cv2
from matplotlib import pyplot as plt
import numpy as np
import mediapipe as mp
from tensorflow.keras.models import load_model 
```

```python
# import dependencies
from IPython.display import display, Javascript, Image
from base64 import b64decode, b64encode
import cv2
import numpy as np
import PIL
import io
import html
import time
```

```python
mpHands = mp.solutions.hands
hands = mpHands.Hands(max_num_hands=3, min_detection_confidence=0.7)
mpDraw = mp.solutions.drawing_utils
```

```python
model_mp= load_model('hand-gesture-recognition-code/mp_hand_gesture')

f = open('hand-gesture-recognition-code/gesture.names', 'r')
classNames = f.read().split('\n')
f.close()
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()

    # Resize image
    img = frame.copy()

    results = model(img)
    
    centers = np.zeros(np.shape(results.xyxy[0])[0])
    
    i = 0
    for (xmin,ymin,xmax,ymax,conf,classe) in results.xyxy[0]:
        frame = cv2.rectangle(frame,(int(xmin),int(ymin)),(int(xmax),int(ymax)),(0,0,255),2) 
        centers[i] = (xmin + xmax)/2
        i = i+1
        

    # BGR 2 RGB
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    x , y, c = img.shape

    # Set flag
    img.flags.writeable = False

    # Detections
    results_mp = hands.process(img)

        # Set flag to true
    img.flags.writeable = True

    # RGB 2 BGR
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    
    if results_mp.multi_hand_landmarks:
        for handslms in results_mp.multi_hand_landmarks: # for each hand
            landmarks = []
            for lm in handslms.landmark:
                lmx = int(lm.x * y)
                lmy = int(lm.y * x)
                landmarks.append([lmx, lmy])
                
            prediction = model_mp.predict([landmarks]) 
            classID = np.argmax(prediction)
            
            min_dist = 1000
            if classID == 1:     #if peace
                mpDraw.draw_landmarks(frame, handslms, mpHands.HAND_CONNECTIONS)
                for i in range(np.size(centers)): 
                    dist = abs(landmarks[0][0] - centers[i])
                    if (dist < min_dist):
                        idx = i
                        min_dist = dist
                xminbox = int(results.xyxy[0][idx][0])
                yminbox = int(results.xyxy[0][idx][1])
                xmaxbox = int(results.xyxy[0][idx][2])
                ymaxbox = int(results.xyxy[0][idx][3])
                frame = cv2.rectangle(frame,(xminbox,yminbox),(xmaxbox,ymaxbox),(0,255,0),2)        
                '''
                    if landmarks[9][1] > xmin and landmarks[9][1] < xmax:
                        frame = cv2.rectangle(frame,(int(xmin),int(ymin)),
                                              (int(xmax),int(ymax)),(255,0,0),2)
                        mpDraw.draw_landmarks(frame, handslms, mpHands.HAND_CONNECTIONS)
                        print(xmin,xmax,landmarks[9][1])
                '''    
    
    print('center',centers)    
    cv2.imshow('Yolov5 Test', frame)
    
    if cv2.waitKey(10) & 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
